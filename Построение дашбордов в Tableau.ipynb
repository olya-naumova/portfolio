# Построение дашбордов в Tableau

TED (от англ. technology, education, design — «технологии, образование, дизайн») — некоммерческий фонд, который проводит популярные конференции. На них выступают специалисты из разных областей и читают лекции на актуальные социальные, культурные и научные темы. 
В проекте были исследованы данные о конференциях и сделаны следующие выводы:
1. История выступлений
* Чаще всего выступления проходили в США -  1885 раз, в Канаде 593 раза, в Великобритании 537 раз, в остальных странах меньше 100 выступлений.
* Популярные категории выступлений: Технологии, Социум, Наука, Культура и Глобальные проблемы.
* К 2020-м, 2021-м годам количество выступлений снизилось и топ популярных тем изменился. Больше выступлений на тему изменений в обществе и в общем о человечестве, скорее всего это связано с ситуации в мире.
2. Тематики выступлений
* Наиболее популярные категории выступлений: Наука, Технология, Культура, Общество и Глобальные проблемы.
* В Канаде темы выступлений Наука и Технологии более популярны чем  в США, где тема Культуры более популярна. 
* Выступлению на тему Общества чаще аплодируют, чем остальным, меньше всего выступлениям на тему Глобальные проблемы. 
* Выступлению Сесиль Ричардс на тему: "Политический прогресс, достигутый женщинами, - И, что дальше?" аплодировали больше всего.
* Больше всего смеются на выступлениях на тему Культуры, меньше всего на тему Глобальных проблем.
* На выступлении Маза Джобрани "Саудовец, индиец и иранец заходят в катарский бар" слушатели смеялись больше всего.
* Очевидной зависимости между длительностью и просмотрами нет, хотя заметно, что выступления длиннее 30 минут просматривают меньше. 
* Выступление Кена Робинсона "Убивают ли школы креативность?" просмотрели больше всего, около 72 млн раз.
* Выступление Рида Хастингса "3 секрета успеха Нетфликса" длилось дольше всех 63 минуты, а просмотрено около 2 млн раз.
3. Авторы выступлений
* Писатель, Художник, Журналист, Автор и Дизайнер - самые популярные виды деятельности у авторов выступлений.
* Обычно у одного автора одно выступление, так же встречаются 2-3 выступления, больше выступлений проводят единицы.
* Дизайнер (Designer) Tom Wujec выступал чаще своих остальных коллег (4 раза), его самое просматривоемое выступление "Постройте башню, соберите команду" собрало 7 018 639 просмотров.
4. Просмотры
* Конференции TED, категорий Наука, Культура, Технологии, Наука и Глобальные проблемы самые просматриваемые.
* Начиная с 2013-го года средние просмотры падают.
